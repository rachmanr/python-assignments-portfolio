```python
import matplotlib.pyplot as plt
import numpy as np


def gradient_descent(f, learning_rate, initial_point):
   
    def deriv(f, base_point):
        return (f(base_point+10**(-5))-f(base_point))/(10**(-5)) ###approximate derivative

    x_coords=[initial_point]
    y_coords=[f(initial_point)] ###store x and y coordinates in lists for the plot

    x_n=(initial_point-learning_rate*deriv(f, initial_point)) ###establish x_0
    
    for i in range(10000):
        x_coords.append(x_n) 
        y_coords.append(f(x_n)) ###add each iteration's x and y values to lists above
        
        x_n=(x_n-learning_rate*deriv(f, x_n)) ###update x_n to x_n+1
        
        if deriv(f, x_n)<10**(-8): ###check if minimum is found before plotting graph
            plot_range=np.linspace(min(x_coords)-0.5, max(x_coords)+0.5, 10000) ###set parameters for graph
            function_range=[f(i) for i in plot_range]
            plt.plot(plot_range, function_range)
            plt.plot(x_coords, y_coords)

            return round(x_coords[-1],3), round(y_coords[-1],3) ###call coordinates of minimum
            
def f(x): ###set test function
    return x**2

learning_rate=0.01 ###set learning rate
initial_point=1 ###set initial point
print(gradient_descent(f, learning_rate, initial_point)) ###call gradient descent function
```
