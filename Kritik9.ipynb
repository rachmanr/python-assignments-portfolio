```python
import sympy as sp

x, y = sp.symbols('x y')
f = sp.exp(x)*sp.sin(y)+y**3

# Differentiate f with respect to x
df_dx = sp.diff(f, x)
# Differentiate f with respect to y
df_dy = sp.diff(f, y)

print('df/dx =', df_dx, 'and df/dy =', df_dy)
```

```python
x, y = sp.symbols('x y') ###symbols
g = x**2*y+x*y**2 ###funcn

dg_dx = sp.diff(g, x) ###partials
dg_dy = sp.diff(g, y)

grad_vec = [dg_dx, dg_dy]

grad_mag = sp.sqrt(1**2 + (-1)**2)

print('gradient vector:', grad_vec, 'and magnitude at (1, -1):', grad_mag)

```

```python
h = sp.ln(x**2+y**2)

dh_dx = sp.diff(h, x) ###1st partials
dh_dy = sp.diff(h, y)

d2h_dx2 = sp.diff(sp.diff(h, x),x) ###2nd partials
d2h_dy2 = sp.diff(sp.diff(h, y),y)
d2h_dxy = sp.diff(sp.diff(h, x),y)
d2h_dyx = sp.diff(sp.diff(h, y),x)

print(d2h_dx2)
print(d2h_dy2)
print(d2h_dxy)
print(d2h_dyx)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
