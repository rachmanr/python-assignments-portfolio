```python
import numpy as np

def normal_density(mean, variance, x):
    nd=(1/np.sqrt(2*np.pi*variance**2))*np.exp((-(x-mean)**2)/(2*variance**2))
    return nd
```

```python
import matplotlib.pyplot as plt

def f(x):
   return normal_density(5,1,x) ###choose random numbers (can't plot with y in terms of three variables)

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), color='chartreuse')

plt.show()
```

```python
n=1000                                 ###loop duration
a=162                                  ###lower integration limit
b=190                                  ###upper integration limit
delx=(b-a)/n                           ###delta x (for ease)
w=0                                    ###initialize before summing
v=7.1                                  ###variance
m=171                                  ###mean

def greg(x):                           ###define function for integration
    return (1/np.sqrt(2*np.pi*v**2))*np.exp((-(x-m)**2)/(2*v**2))

for i in range(0,n+1):                 ###implementing Simpson's method for numerically solving integrals
    x_i=a+i*delx                       ###calculate x value for each iteration
    y_i=greg(x_i)                      ###calculate corresponding y value

     ###establish placement of three possible coefficients for each term in Simpson's method
    if i==0 or i==n+1:
        w+=y_i
    elif i%2==1 and not i==n+1:
        w+=4*y_i
    elif i%2==0 and not i==0 and not i==n+1:
        w+=2*y_i
        
s=(delx/3)*w                           ###calculate final integral
print(s)                               ###call integral
        


```
